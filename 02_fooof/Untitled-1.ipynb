```python
import numpy as np
from scipy.signal import welch
from specparam import SpectralGroupModel

np.random.seed(0)
freqs, psds = welch(np.random.randn(5, 2000), fs=200, nperseg=400)

fg = SpectralGroupModel(max_n_peaks=3)

fg.fit(freqs, psds, freq_range=(2, 40), n_jobs=-1)
print(fg.get_metrics("error", "mae"))     # → [nan nan nan nan nan]

fg.fit(freqs, psds, freq_range=(2, 40), n_jobs=1)
print(fg.get_metrics("error", "mae"))     # → [0.087  0.122  0.087  ...]
```

```python
import numpy as np
from scipy.signal import welch
from specparam import SpectralGroupModel

np.random.seed(0)
freqs, psds = welch(np.random.randn(5, 2000), fs=200, nperseg=400)

fg = SpectralGroupModel(max_n_peaks=3)

# --- parallel (broken) ---
fg.fit(freqs, psds, freq_range=(2, 40), n_jobs=-1)
print("=== n_jobs=-1 ===")
print("aperiodic:", fg.get_params("aperiodic"))   # ✓ correct
print("peaks:    ", fg.get_params("peak"))         # ✗ all NaN (CF/PW/BW)
print("mae:      ", fg.get_metrics("error", "mae")) # ✗ all NaN

# --- serial (correct) ---
fg.fit(freqs, psds, freq_range=(2, 40), n_jobs=1)
print("\n=== n_jobs=1 ===")
print("aperiodic:", fg.get_params("aperiodic"))   # ✓ correct
print("peaks:    ", fg.get_params("peak"))         # ✓ correct
print("mae:      ", fg.get_metrics("error", "mae")) # ✓ correct
```
